```python
import requests
import zipfile
import os
```

```python


# Step 1: Download the ZIP file
url = "https://dimensions.figshare.com/ndownloader/files/30718538"
zip_path = "/content/file.zip"

response = requests.get(url)
with open(zip_path, "wb") as file:
    file.write(response.content)

```

```python
# Step 2: Extract the ZIP file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("/content/files")

# Step 3: List the files in the extracted directory to verify
extracted_files_path = "/content/files"
extracted_files = os.listdir(extracted_files_path)
print(extracted_files)
```
